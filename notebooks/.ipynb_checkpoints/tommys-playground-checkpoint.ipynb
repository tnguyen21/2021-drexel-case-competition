```python
import pandas

filepath = '../datasets/urban_menu_data_raw.csv'
raw_menu_data = pandas.read_csv(filepath, encoding="unicode-escape")
print("len of raw menu data:", len(raw_menu_data))

# Cleaning Data

cleaned_df = raw_menu_data.fillna(0)
cleaned_df = cleaned_df.drop_duplicates(subset=['item_name'])
# cleaned_df.to_csv("../datasets/urban_menu_data_duplicates_removed.csv", index=False)
print("len after duplicates dropped", len(cleaned_df))
```

```python
# Some basic stats

display(cleaned_df)

print(cleaned_df['is_locally_grown'].value_counts())
print(cleaned_df['is_organic'].value_counts())
print(cleaned_df['is_vegetarian'].value_counts())
print(cleaned_df['is_vegan'].value_counts())
```
